```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

%matplotlib widget
```

```python
df = pd.read_excel('../forest-monitoring/results.xlsx', sheet_name='DBH')
ref = df['Ground Truth DBH (cm)'][0:80].to_numpy()
pred = df['Measured DBH (cm)'][0:80].to_numpy()

# remove the NaN values
ref = ref[~np.isnan(ref)]
pred = pred[~np.isnan(pred)]
```

```python
np.min(pred)
```

##### 1. Regression Plot

```python
# obtain m (slope) and b (intercept) for the linear regression line
m, b = np.polyfit(pred, ref, 1)

plt.figure()
plt.grid()
plt.scatter(pred, ref, color='blue')
plt.plot(ref, m*ref + b, color='red')
plt.xlabel('Reference Values (cm)')
plt.ylabel('Extracted Values (cm)')
plt.title('Regression Plot')    
plt.show()
```

##### 2. Regression Plot with Confidence Intervals

```python
sns.regplot(x=pred, y=ref)
plt.show()
```

```python
df = pd.read_excel('../forest-monitoring/results.xlsx', sheet_name='CD & TH')
cd_ref = df['Ground Truth CD (cm)'][0:10].to_numpy()
cd_pred = df['Extracted CD (cm)'][0:10].to_numpy()
th_ref = df['Ground Truth TH (cm)'][0:10].to_numpy()
th_pred = df['Extracted TH (cm)'][0:10].to_numpy()

# remove the NaN values
cd_ref = cd_ref[~np.isnan(cd_ref)]
cd_pred = cd_pred[~np.isnan(cd_pred)]
th_ref = th_ref[~np.isnan(th_ref)]
th_pred = th_pred[~np.isnan(th_pred)]
```

```python
# obtain m (slope) and b (intercept) for the linear regression line
m, b = np.polyfit(cd_pred, cd_ref, 1)

plt.figure()
plt.scatter(cd_pred, cd_ref, color='blue')
plt.plot(cd_ref, m*cd_ref + b, color='red')
plt.xlabel('Reference Values (cm)')
plt.ylabel('Extracted Values (cm)')
plt.title('Regression Plot')    
plt.show()
```

```python
# obtain m (slope) and b (intercept) for the linear regression line
m, b = np.polyfit(th_pred, th_ref, 1)

plt.figure()
plt.scatter(th_pred, th_ref, color='blue')
plt.plot(th_ref, m*th_ref + b, color='red')
plt.xlabel('Reference Values (cm)')
plt.ylabel('Extracted Values (cm)')
plt.title('Regression Plot')    
plt.show()
```
